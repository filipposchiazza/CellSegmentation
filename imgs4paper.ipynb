```python
import os
import sys
import pickle
import json
sys.path.append('./modules')
from modules.utils import draw_countour
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load configuraion file
with open('modules/config.json') as f:
    config = json.load(f)
```

```python
real_imgs_dir = config['dataset']['img_dir']
cond_syn_img_dir = config['dataset']['cond_syn_img_dir']
uncond_syn_img_dir = config['dataset']['uncond_syn_img_dir']
four_channels_img_dir = config['dataset']['4channels_img_dir']
assert os.path.exists(real_imgs_dir)
assert os.path.exists(cond_syn_img_dir)
assert os.path.exists(uncond_syn_img_dir)
assert os.path.exists(four_channels_img_dir)
```

```python
real_imgs = ['CTR_80.png', 'CTR_12946.png', 'CTR_261.png', 'CTR_178.png']
cond_syn_imgs = ['gen_img_535.png', 'gen_img_1442.png', 'gen_img_707.png', 'gen_img_692.png']
uncond_syn_imgs = ['gen_img_95.png', 'gen_img_103.png', 'gen_img_304.png', 'gen_img_402.png']
uncond_4channels_imgs = ['gen_img_95.png', 'gen_img_1030.png', 'gen_img_2421.png', 'gen_img_402.png']

# plot the images in a grid 4x4
fig, axs = plt.subplots(4, 4, figsize=(15, 8))
for i in range(4):
    img = plt.imread(os.path.join(real_imgs_dir, real_imgs[i]))
    axs[i, 0].imshow(img)
    axs[i, 0].axis('off')
    if i == 0:
        axs[i, 0].set_title('Real')

    img = plt.imread(os.path.join(cond_syn_img_dir, cond_syn_imgs[i]))
    axs[i, 1].imshow(img)
    axs[i, 1].axis('off')
    if i == 0:
        axs[i, 1].set_title('Mask Conditioned')

    img = plt.imread(os.path.join(uncond_syn_img_dir, uncond_syn_imgs[i]))
    axs[i, 2].imshow(img)
    axs[i, 2].axis('off')
    if i == 0:
        axs[i, 2].set_title('Unconditioned (image only)')

    img = plt.imread(os.path.join(four_channels_img_dir, uncond_4channels_imgs[i]))
    axs[i, 3].imshow(img)
    axs[i, 3].axis('off')
    if i == 0:
        axs[i, 3].set_title('Unconditioned (image + mask)')

```
